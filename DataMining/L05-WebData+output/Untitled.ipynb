## 股票数据分析

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv('data/stock_px.csv', index_col='date')
data
```

```python
data.index = pd.to_datetime(data.index)
data
```

```python
plt.rc('figure', figsize=(12, 6))
data.plot.line()
```

```python
data[['AAPL', 'IBM', 'MSFT', 'GE', 'AA', 'JNJ', 'PEP']].plot.line()
```

```python
data['AAPL'].describe()
```

```python
data.corr()
```

```python
data
```

```python
data['AAPL'].tail()
```

```python
df = pd.DataFrame(data['AAPL'], index=data.index, columns=['AAPL'])
df.tail()
```

```python
df['Return'] = 0.0
for i in range(1, len(df)):
#     df['Return'][i] = np.log(df['AAPL'][i] / df['AAPL'][i-1])
#     print(np.log(df['AAPL'][i] / df['AAPL'][i-1]))
    df['Return'][i] = np.log(df['AAPL'][i] / df['AAPL'][i-1])
df.tail()
```

```python
df['Return2'] = np.log(df['AAPL'] / df['AAPL'].shift(1))
df
```

```python
# del df['Return2']
df
```

```python
df[['AAPL', 'Return']].plot(subplots=True, style=['b', 'r'], figsize=(8, 5))
```

```python
df['42d'] = df['AAPL'].rolling(window=42, center=False).mean()
df['252d'] = df['AAPL'].rolling(window=252, center=False).mean()
df['512d'] = df['AAPL'].rolling(window=512, center=False).mean()
```

```python
df.head()
```

```python
df.tail()
```

```python
df[['AAPL', '42d','252d', '512d']].plot.line()
```

```python
out_file = open('data/aapl.csv', 'w')
df.to_csv(out_file)
out_file.close()
```
