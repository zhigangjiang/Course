```python
import pandas as pd
import numpy as np
df = pd.DataFrame(data = [1, 2, 3, 0, 1, 2], columns=["reli"])
df = df.reset_index()
df = df.rename(columns = {"index":"i"})
df["i"] += 1
df["log2(i+1)"] = np.log2(df["i"]+1)
df["reli /log2(i+1)"] = df["reli"] /df["log2(i+1)"]
df
```

```python
CG = df["reli"].sum()
CG
```

```python
DCG = df["reli /log2(i+1)"].sum()
DCG
```
